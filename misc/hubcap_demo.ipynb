# Git Status for Multiple Repos

This section shows how to get the git status of all git repositories under a root directory (recursively).

````python
from pathlib import Path
from hubcap.util import git
from typing import Dict, List


def find_git_repos(rootdir: str) -> List[Path]:
    """
    Find all git repositories under a root directory (recursively).
    
    Args:
        rootdir: Root directory to search for git repositories
        
    Returns:
        List of Path objects pointing to git repository directories

    Note: Could also use gita (https://pypi.org/project/gita/) for this
    """
    rootdir = Path(rootdir).expanduser().resolve()
    git_repos = []
    
    for git_dir in rootdir.rglob('.git'):
        if git_dir.is_dir():
            # The parent of .git is the repo directory
            git_repos.append(git_dir.parent)
    
    return sorted(git_repos)


def get_repo_status(repo_path: Path) -> Dict[str, str]:
    """
    Get git status for a single repository.
    
    Args:
        repo_path: Path to the git repository
        
    Returns:
        Dict with repo_path and status information
    """
    try:
        status = git('status --short', work_tree=str(repo_path))
        # Also get branch info
        branch = git('branch --show-current', work_tree=str(repo_path))
        
        return {
            'path': str(repo_path),
            'branch': branch,
            'status': status if status else 'clean',
            'has_changes': bool(status)
        }
    except Exception as e:
        return {
            'path': str(repo_path),
            'branch': 'error',
            'status': f'Error: {e}',
            'has_changes': False
        }


def get_all_repos_status(rootdir: str) -> List[Dict[str, str]]:
    """
    Get git status for all repositories under a root directory.
    
    Args:
        rootdir: Root directory to search for git repositories
        
    Returns:
        List of dicts containing status info for each repository
    """
    repos = find_git_repos(rootdir)
    return [get_repo_status(repo) for repo in repos]
````

## Alternative: Simpler one-liner approach

If you just want a quick overview without all the functions:

````python
# Quick one-liner to get status of all repos
from pathlib import Path
from hubcap.util import git

rootdir = Path('~/Dropbox/py/proj/i').expanduser()

for repo in sorted(rootdir.rglob('.git')):
    if repo.is_dir():
        repo_path = repo.parent
        try:
            status = git('status --short', work_tree=str(repo_path))
            branch = git('branch --show-current', work_tree=str(repo_path))
            if status:  # Only print if there are changes
                print(f"\n{repo_path.name} ({branch}):")
                print(status)
        except Exception as e:
            print(f"\n{repo_path.name}: Error - {e}")
````
